```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:/Users/HP/Downloads/heart.csv")
```

```python
df
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=43)
```

```python
data = DecisionTreeClassifier()
```

```python
data.fit(x_train,y_train)
```

```python
y_pred = data.predict(x_test)
```

```python
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f" Decision Tree Accuracy: {accuracy:.2f}")
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
parameter={
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Classifier=DecisionTreeClassifier()
cv=GridSearchCV(Classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(x_train,y_train)
```

```python
("Best Hyperparameters:", cv.best_params_)
```

```python
y_predict=cv.predict(x_test)
y_predict
```

```python
accuracy = accuracy_score(y_test, y_predict)
print(f'Accuracy: {accuracy:.2f}')
```

```python
report = classification_report(y_test, y_predict)
print("Classification Report:\n", report)
```

```python
cm = confusion_matrix(y_test, y_predict)
```
